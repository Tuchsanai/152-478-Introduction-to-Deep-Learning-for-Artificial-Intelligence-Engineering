```python
import cv2
import matplotlib.pyplot as plt
from skimage import io


'''
สมาชิค
1.ศุภชัย    เตชะธนไพบูลย์   6304200016
2.ชาคริต   เอี่ยมสกุล       6223200019
3.ณัฐวุฒิ    แก้วประไพ      6223200012
4.ณัฐพงศ์   นาคทรัพย์       6223200010
5.รวิพล    ขำเมืองน้อย     6221200004
6.ฤกษ์ดี    ช่างแกะ        6223200003
7.กมลเทพ  จิตรณรงค์       6223200013

'''
#import รูป & convert into RGB scale
orlottery = cv2.imread("/content/drive/MyDrive/หวย.JPG")  
orlottery.shape

#convert color scale to normal 
orlottery = cv2.cvtColor(orlottery, cv2.COLOR_BGR2RGB)

#plot orlottery
plt.figure(figsize=(10,10))
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(orlottery)

h,w,c = orlottery.shape
print('h=',h,'w=',w,'c=',c)

image = orlottery.copy()
height, width = image.shape[:2] #import ข้อมูล 2 ตัวแรก
rotation_matrix = cv2.getRotationMatrix2D((0.5*width, 0.5*height), 205.0, 1.2) #set config; (0.5*width, 0.5*height) = rotate about origin, 205.0 = rotating angle, 1.2 magnifying scale
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height)) #rotate image

#print rotated image
plt.figure(figsize=(10,10))
plt.subplot(1, 3, 2)
plt.title("Rotated")
plt.imshow(rotated_image)

#crop image
rotated_image.shape
cropimg = rotated_image[480:900,180:1050,:]
cropimg.shape

#print cropped image
plt.figure(figsize=(10,10))
#plt.subplot(1, 3, 3)
plt.title("Complete")
plt.imshow(cropimg)





'''Group Photo Assignment'''



#import sky picture from url
url_sky = "https://raw.githubusercontent.com/Tuchsanai/152-478-Introduction-to-Deep-Learning-for-Artificial-Intelligence-Engineering/main/02_Crash-NumPy/skyimg.jpg"
sky      = io.imread(url_sky) 
io.imsave('sky.jpg',sky) #save sky image

resized_sky = cv2.resize(sky,(2000,1000) , interpolation = cv2.INTER_AREA) #(column,row)
            
#Image 1
person1 = io.imread("/content/drive/MyDrive/ศุภชัย 6304200016.jpg")
resized_person1 = cv2.resize(person1,(400,300) , interpolation = cv2.INTER_AREA)                    #resize image, (400,300) column,row
cv2.putText(resized_person1,'Suppachai 6304200016',(10,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,100),4)     #Add text

#Image2
person2 = io.imread("/content/drive/MyDrive/ชาคริต เอี่ยมสกุล 6223200019.jpg")
resized_person2 = cv2.resize(person2,(400,300) , interpolation = cv2.INTER_AREA)                    #resize image, (400,300) column,row
cv2.putText(resized_person2,'Chakrid 622320001',(30,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),4)     #Add text

#Image3
person3 = io.imread("/content/drive/MyDrive/ณัฐวุฒิ แก้วประไพ 6223200012.jpg")
resized_person3 = cv2.resize(person3,(400,300) , interpolation = cv2.INTER_AREA)                    #resize image, (400,300) column,row
cv2.putText(resized_person3,'Nuttawud 6223200012',(30,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),4)     #Add text

#Image4
person4 = io.imread("/content/drive/MyDrive/นายณัฐพงศ์ นาคทรัพย์ 6223200010.jpg")
resized_person4 = cv2.resize(person4,(400,300) , interpolation = cv2.INTER_AREA)                    #resize image, (400,300) column,row
cv2.putText(resized_person4,'Nuttapong 6223200010',(30,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),4)     #Add text

#Image5
person5 = io.imread("/content/drive/MyDrive/moss..jpg")
resized_person5 = cv2.resize(person5,(400,300) , interpolation = cv2.INTER_AREA)                    #resize image, (400,300) column,row
cv2.putText(resized_person5,'Rawipon 6221200004',(30,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),4)     #Add text

#Image6
person6 = io.imread("//content/drive/MyDrive/Roekdee 622320003.jpg")
resized_person6 = cv2.resize(person6,(400,300) , interpolation = cv2.INTER_AREA)                    #resize image, (400,300) column,row
cv2.putText(resized_person6,'Roekdee 622320003',(30,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,255),4)     #Add text

#Image7
person7 = io.imread("/content/drive/MyDrive/Kamonthep 6223200013.jpg")                    
height, width = person7.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((0.5*width, 0.5*height), 270.0, 1)
person7 = cv2.warpAffine(person7, rotation_matrix, (width, height))
resized_person7 = cv2.resize(person7,(400,300) , interpolation = cv2.INTER_AREA)
cv2.putText(resized_person7,'Kamonthep 6223200013',(30,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),4)  

#Add Group Name
cv2.putText(resized_sky,'Group 4 :)',(200,600),cv2.FONT_HERSHEY_SIMPLEX,10,(255,170,180),8) #(colum,row)


resized_sky[0:300,0:400,:] = resized_person1 #[row,colmn]
resized_sky[0:300,530:930,:] = resized_person2
resized_sky[0:300,1060:1460,:] = resized_person3
resized_sky[0:300,1600:2000,:] = resized_person4
resized_sky[700:1000,0:400,:] = resized_person5
resized_sky[700:1000,800:1200,:] = resized_person6
resized_sky[700:1000,1600:2000,:] = resized_person7 #rotated

plt.figure(figsize=(20, 20))
plt.title("Group 4 Members")
plt.imshow(resized_sky)
```
