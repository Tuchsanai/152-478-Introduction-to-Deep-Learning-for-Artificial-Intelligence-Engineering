```python
import pandas as pd
import numpy as np
import random


url    = "https://home.sis.siam.edu/registrar/student_inclass_all.asp?backto=teach_time&option=1&classid=208419&courseid=14949&acadyear=2564&semester=2&normalURL=f%5Fcmd%3D2%26officercode%3D53113029%26officerid%3D20710%26officername%3D%25BC%25C8%252E%25B4%25C3%252E%2B%25B7%25D1%25C8%25B9%25D1%25C2%2B%25BE%25C5%25CD%25C2%25CA%25D8%25C7%25C3%25C3%25B3%26acadyear%3D2564%26rnd%3D44585%2E5031944444%26firstday%3D24%2F1%2F2565%26semester%3D2&avs652493303=9"
List = pd.read_html(url)

df = List[3]
df =  df.loc[3:,0:2].reset_index( drop = True).drop(0,axis=1)
df['No']     = df.index + 1

df0= df.rename(columns={1:'id',2:'Name'})
df0 =df0[['No','id','Name']]
df0['id']= df0['id'].astype('string')
df = df0

df
```

```python
df_regis = pd.read_csv('data.csv')
df_regis = df_regis[['ชื่อ-สกุล', 'รหัสนักศึกษา', 'line id','Email address','ชื่อ  ที่แสดงในกลุ่ม Line', 'เบอร์โทร']]
df_regis = df_regis.rename({"ชื่อ-สกุล":'Name','รหัสนักศึกษา':'id'},axis=1 )
df_regis.fillna(0,inplace=True)
df_regis['id'] = df_regis['id'].astype('int').astype('str')
df_regis = df_regis.set_index('Name')


for index, x in df.iterrows() :
    name = x['Name'] 
    if name in df_regis.index.tolist() :
      df_regis.loc[name,'id'] = x['id']



df_regis = df_regis.reset_index()

df_regis
```

```python
L_not_assign = []

for x in df['Name'] :
    if x in df_regis['Name'].tolist() :
        pass
    else :
        L_not_assign.append(x)
        print(x)

```

```python
groups         = 5


Header = ['นางสาวมนปพร ชัยวาณิชยา',
          'นายศุภชัย เตชะธนไพบูลย์',
          'นายนนทวัฒน์ บุญเป็ง'
          ,'AJ P',
          'Arm'
          ]


dfx          = pd.merge(df,df_regis,left_on="Name", right_on="Name",how='outer')
dfx.drop('id_y',axis=1,inplace=True)     
dfx['groups'] = -1
dfx = dfx.rename({"id_x":'Id'},axis=1 )

random.shuffle(Header)


dfx = dfx.set_index('Name')          

for i,name in enumerate(Header) :
   dfx.loc[name,'groups']  = (i % groups) +1


for i,name in enumerate(L_not_assign) :
   dfx.loc[name,'groups']  = 10


dfx = dfx.reset_index()
```

```python
G      = dfx['groups'].value_counts()
members = G[-1]//groups
remains = G[-1]% groups



G = G.sort_values()

for x in G.keys() :
    if x!= 10 and x!= -1 :
        G[x]=G[x]+ members


for x in G.keys() :
    if remains >0 and x!= 10 and x!= -1 :
        G[x]=G[x]+1
        remains =remains-1

G= G.drop(-1,axis=0)
G= G.drop(10,axis=0)



D = []

for x in G.keys() :
    n0 = dfx['groups'].value_counts()[x]
    if x!= 10 and x!= -1  :  
      D += (G[x]-n0) *[x]

random.shuffle(D)


G
```

```python
I = dfx[dfx['groups']==-1].index.tolist()
dfx.loc[I,'groups'] =D
dfx = dfx.sort_values(by=['No'])


```
