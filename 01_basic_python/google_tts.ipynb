```python
from IPython.display import Audio
import gtts
from gtts import gTTS


def repr_html(self):
  import io
  if not hasattr(self, 'data'):
    fp = io.BytesIO()
    self.write_to_fp(fp)
    self.data = fp.getvalue()
  audio = Audio(self.data, autoplay=True)
  return audio._repr_html_()

gTTS._repr_html_ = repr_html
```
