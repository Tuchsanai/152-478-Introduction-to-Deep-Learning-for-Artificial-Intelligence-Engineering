```python
import pytesseract
from pytesseract import Output
import cv2
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
url = "img_vernier/626520.jpg"
img = io.imread(url)
plt.figure(figsize=(10,10))
plt.imshow(img)
```

# Crop Image of Interest

```python
x = img[295:500, 855:920]
x_hsv = cv2.cvtColor(x, cv2.COLOR_RGB2HSV)

plt.subplot(2,1,1)
plt.imshow(x)
plt.subplot(2,1,1)
plt.imshow(x)
```

# find Histogram

```python
hist0 = cv2.calcHist([x_hsv], [0], mask = None, histSize = [256], ranges = [0, 256])
hist1 = cv2.calcHist([x_hsv], [1], mask = None, histSize = [256], ranges = [0, 256])
hist2 = cv2.calcHist([x_hsv], [2], mask = None, histSize = [256], ranges = [0, 256])

plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
plt.plot(hist0)
plt.xlim([150, 200])
plt.subplot(3,1,2)
plt.plot(hist1)
plt.xlim([80, 255])
plt.subplot(3,1,3)
plt.plot(hist2)
plt.xlim([0, 255])
```

```python
  
# define range of blue color in HSV
low_red = np.array([168, 100, 210], dtype=np.uint8)
high_red = np.array([175, 220,255], dtype=np.uint8)


hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

# Threshold the HSV image to get only blue colors
mask = cv2.inRange(hsv, low_red, high_red)



# Bitwise-AND mask and original image
res = cv2.bitwise_and(img,img, mask= mask)

plt.figure(figsize=(50,50))
plt.subplot(1,3,1)
plt.imshow(img)
plt.subplot(1,3,2)
plt.imshow(mask)
plt.subplot(1,3,3)
plt.imshow(res)
```

```python
ximg = img.copy()

def find_max_area(contour):
    max_area = 0
    max_contour = None
    second_max_contour = None
    for cnt in contour:
        area = cv2.contourArea(cnt)
        if area > max_area:
            max_area = area
            second_max_contour = max_contour
            max_contour = cnt
    return max_contour, second_max_contour


contour,hierarchy = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

if len(contour) > 0 :
   ctn,ctn2 =find_max_area(contour)  
   x,y,w,h = cv2.boundingRect(ctn)
   mask_img1 = mask[y:y+h,x:x+w]
   imask_img1 = cv2.bitwise_not( mask_img1)
   img1     = img[y:y+h,x:x+w]
   plt.imshow( imask_img1)
```

```python
contour,hierarchy = cv2.findContours(imask_img1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
ctn,ctn2 =find_max_area(contour)  
  
```

```python
x,y,w,h = cv2.boundingRect(ctn)
yimg    = img1[y:y+h,x:x+w,:]
plt.imshow(yimg)
```
