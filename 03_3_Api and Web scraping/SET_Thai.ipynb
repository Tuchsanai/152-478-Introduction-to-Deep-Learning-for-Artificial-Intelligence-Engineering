```python
import time
from bs4 import BeautifulSoup
import requests
import pandas as pd

async def get_info(sname) :
    
    xurl = f'https://www.set.or.th/set/factsheet.do?symbol={sname}&ssoPageId=3&language=th&country=TH'
    html =  requests.get(xurl)
    
    bs = BeautifulSoup(html.content, "html.parser")
    result_findall = bs.find_all('table',{"border":"0" ,"cellpadding":"0", "cellspacing":"0","class":"factsheet"} )

    text = None
    Price = None
    
    for tag_table in result_findall :

        if 'Price' in tag_table.get_text() :
            td_tag = tag_table.find('td',{"class":"factsheet"})
            if td_tag :
                Price = td_tag.get_text()
                break

    
    for tag_table in result_findall :

        if 'ลักษณะธุรกิจ' in tag_table.get_text() :
            td_tag = tag_table.find('td',{"class":"factsheet"})
            if td_tag :
                text = td_tag.get_text()
                break
    return text, Price   


stocks = ['BEM','PTT','AIMIRT','DTAC','BDMS','ADVANC','BBL','HMPRO','INTUCH','TISCO','TOP','SAWAD']
texts   = []
prices  = []

for stock_name in stocks :
    info, Price = await get_info(stock_name)
    texts.append(info)
    prices.append(Price)
    print(stock_name,':',Price,':',info)
    
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', 199)  # or 199

df = pd.DataFrame({'name':stocks,'Price':prices,'info':texts})
df
```
